## LRG Catalogue Production

Let's see how the halo $V_{peak}$ distribution.

```python
data = halo[1].data
vpeak = data['vpeak']
print(vpeak.shape)
```

```python
plt.xlabel('$V_{peak}$(km/s)')
plt.ylabel('# of halos')
plt.hist(vpeak,bins=50)#,range=(120,600))
plt.title('halo catalogue vpeak distribution')
plt.yscale('log');
```

In order to see the difference between the original catalogue and the scattered catalogue, we sort them by $V_{peak}$ and $V_{peak}^{scat}$, respectively. 

```python
for i in [397.69999,397.699996,397.699997,397.699999]:
    print('vpeak >',i,'-',len(vpeak[vpeak>i]))
```

For the original version $V_{peak}$, the number of haloes that has $V_{peak}>397.7$ is around the number of eBOSS LRG (5468750), meaning $V_{peak}=397.7km/s$ is threshold of the LRG catalogue. But there is something strange happened between 397.69999 and 397.699999. 

```python
time_start=time.time()
sort_raw = data[data['vpeak'].argsort()]
LRGorg = sort_raw[::-1][:5468750]
time_end=time.time()
print('time cost',time_end-time_start,'s')
```

```python
org_tmp = Table([LRGorg['X'],LRGorg['Y'],LRGorg['Z'],LRGorg['vx'],LRGorg['vy'],LRGorg['vz']], names=('x','y','z','vx','vy','vz'))
ascii.write(org_tmp,path+"LRGorg.dat", delimiter='\t')
```

For the scattered version, the $V_{peak}^{scat}=(1+N(0,\sigma_{HAM}))V_{peak}$ where we take scatter width $\sigma_{HAM}=0.31$ from S. RodrÃ­guez-Torres et al. (2016). Our pre-processing step keep haloes with $V_{peak}>120 km/s$. In principle, we should make sure that the haloes with the smallest $V_{peak}^{crit}\approx 120km/s$ could hardly be scattered to the range where they can have a galaxy inside, i.e., $V_{peak}^{scat,crit}>400km/s$. It means that $V_{peak}^{scat,crit}>(1+5\sigma_{HAM})V_{peak}^{crit}$. Here $400km/s>(1+5*0.31)*120km/s$, the thresholds and $\sigma_{HAM}$ values are satisfactory.

```python
np.random.seed(0)
time_start=time.time()
datac = np.copy(data)
datac['vpeak'] *= 1+np.random.normal(scale=0.31,size=len(datac['vpeak']))
sort_scat = datac[datac['vpeak'].argsort()]
sort_scatadd = data['vpeak'][datac['vpeak'].argsort()]
LRGscat = sort_scat[::-1][:5468750]
scatorg = sort_scatadd[::-1][:5468750]
time_end=time.time()
print('time cost',time_end-time_start,'s')
```

```python
scat_tmp = Table([LRGscat['X'],LRGscat['Y'],LRGscat['Z'],LRGscat['vx'],LRGscat['vy'],LRGscat['vz']], names=('x','y','z','vx','vy','vz'))
ascii.write(scat_tmp,path+"LRGscat.dat", delimiter='\t')
```

```python
fig=plt.figure(figsize=(14,5))
N =[1,2]
edge=[1,2]
for i,j in zip([scatorg,LRGscat['vpeak']],range(2)):
    ax = plt.subplot2grid((1,2),(0,j))
    ax.set_xlabel('$V_{peak}$(km/s)')
    ax.hist(i,bins=50)
    N[j],edge = np.histogram(i,bins=50)
    ax.set_yscale('log')
    print(edge[N[j].argmax()])
    if j==1:
        plt.title('vpeak_scatter value distribution')
        ax.set_ylabel('# of LRGs')
    if j==0:
        plt.title('vpeak_scatter original vaule distribution')
        ax.set_ylabel('# of haloes')
```

The selection of scattered samples was wrong because I erroneously chose the smallest ones instead of the largest ones. After the correction, we could see that the number of selected samples increases exponentially as the $V_{peak}^{scat}$ becomes smaller until ~440 km/s. While their corresponding original $V_{peak}$ has an exponential part and a turnover point ~350 km/s. The turning point appears because the scattering allows some light haloes to have an LRG, but the pdf follows Gaussian distribution. Meanwhile, if the haloes are heavy enough, it is difficult to scatter them out of the catalogue. Their vacancies in the $V_{peak}^{scat}$ bins will be filled by other scattered haloes. So the $V_{peak}^{scat}$ still has an exponential part. 

The plot above is the probability for a halo to have an LRG inside. This distribution is within our expectation. But we are not sure the intermediate part is Guassian or not. 

We need to vary the $\sigma_{HAM}$ to see its effect. So we choose $\sigma_{HAM}=0.1,0.2$ and $0.4$, respectively and repeat the tests. Our expectation is that the smaller the $\sigma_{HAM}$ is, the more similar it will be compared to the 2PCF from the original catalogue.

```python
np.random.seed(0)
time_start=time.time()
datac1 = np.copy(data)
datac1['vpeak'] *= 1+np.random.normal(scale=0.1,size=len(datac1['vpeak']))
sort_scat1 = datac1[datac1['vpeak'].argsort()]
LRGscat1 = sort_scat1[::-1][:5468750]
sort_scatadd1 = data['vpeak'][datac1['vpeak'].argsort()]
scatorg1 = sort_scatadd1[::-1][:5468750]
scat_tmp1 = Table([LRGscat1['X'],LRGscat1['Y'],LRGscat1['Z'],LRGscat1['vx'],LRGscat1['vy'],LRGscat1['vz']], names=('x','y','z','vx','vy','vz'))
ascii.write(scat_tmp1,path+"LRGscat1.dat", delimiter='\t')
time_end=time.time()
print('time cost',time_end-time_start,'s')
```

```python
n,bins=np.histogram(vpeak,bins=50,range=(120,2500))
n0,bins0=np.histogram(scatorg,bins=50,range=(120,2500))
fig,ax=plt.subplots()
ax.plot(bins[:-1],n0/n,c='b',label='$\sigma=0.31$')

n1,bins1=np.histogram(scatorg1,bins=50,range=(120,2500))
#fig,ax=plt.subplots()
ax.set_xlabel('$V_{peak}$(km/s)')
ax.plot(np.linspace(120,2500,10),np.ones(10),'k--')
ax.plot(bins[:-1],n1/n,c='r',label='$\sigma=0.1$')
ax.set_xlim(2500,120)
plt.legend(loc=3)
ax.set_ylabel('prob. of having an LRG');
```

## 2PCF Differece

```python
cfpath =  '/global/cscratch1/sd/jiaxi/master/2PCFout/'+'0211_LRG_real/'#'0209/
```

```python
from astropy.io import ascii
org = Table.read(cfpath+'LRGorg.dat',format="ascii")
scat= Table.read(cfpath+'LRGscat.dat',format="ascii")
scat1= Table.read(cfpath+'LRGscat1.dat',format="ascii")
scat2= Table.read(cfpath+'LRGscat2.dat',format="ascii")
scat3= Table.read(cfpath+'LRGscat3.dat',format="ascii")
```

```python
# 0209
fig=plt.figure(figsize=(18,5))
for i,j,name in zip(['col2','col3','col4'],range(3),['monopole','quadrapole','hexadecapole']):
    ax = plt.subplot2grid((1,3),(0,j))
    ax.plot(org['col1'],org['col1']**2*org[i],c='k',label='original')
    ax.plot(scat1['col1'],scat3['col1']**2*scat1[i],c='purple',alpha=0.5,label='$\sigma=0.1$')
    ax.plot(scat2['col1'],scat1['col1']**2*scat2[i],c='c',label='$\sigma=0.2$')
    ax.plot(scat['col1'],scat['col1']**2*scat[i],c='r',label='$\sigma=0.31$(paper)')
    ax.plot(scat3['col1'],scat2['col1']**2*scat3[i],c='orange',alpha=0.5,label='$\sigma=0.4$')
    
    plt.legend(loc=0)
    plt.xlabel('d_cov (Mpc $h^{-1}$)')
    plt.ylabel('d_cov^2 * $\\xi$')
    plt.title('2PCF '+name+' comparison')
```

In the two-point correlation function monopole plot, the galaxy catalogues are systematically smaller than the original catalogue for positive values. This is because the LRGs generated from haloes are diluted bacause the scattering rearranges their neibours to far-away haloes or even exclude them. So the galaxy correlation function is smaller in positive side. For the negative values,($d_{com}>110$Mpc $h^{-1}$), haloes are more likely to correlate with voids since the selected haloes are massive and more probable to be close to each other, so the halo signals are smaller. $\xi$ difference caused by $\sigma_{HAM}$ will be explained later.


Our plots are similar to D. Eisenstein et al.(2005) fig, but systematically smaller. This is because their results are conducted in 0.16<z<0.47, but our catalogue is in z=0.56. There are more structures in lower redshift, i.e., there are more galaxies. So their signals are larger than us.

For quadrupole and hexadecapole, since we didn't include the redshift space disturtion, so in principle they should be 0. 

```python
# 0211
fig=plt.figure(figsize=(12,3))
for i in range(2):
    ax = plt.subplot2grid((1,2),(0,i))
    ax.plot(scat1['col1'],scat1['col2']/org['col2'],c='purple',alpha=0.5,label='$\sigma=0.1$')
    ax.plot(scat2['col1'],scat2['col2']/org['col2'],c='c',label='$\sigma=0.2$')
    ax.plot(scat['col1'],scat['col2']/org['col2'],c='r',label='$\sigma=0.31$')
    ax.plot(scat3['col1'],scat3['col2']/org['col2'],c='orange',alpha=0.5,label='$\sigma=0.4$')
    #plt.plot(scat2['col1'],np.ones_like(scat2['col1']),c='k')
    plt.legend(loc=0)
    plt.xlabel('d_cov (Mpc $h^{-1}$)')
    plt.ylabel(' $\\xi$ ratio')
    plt.title('2PCF monopole scatter/original')
    if i==1:
        plt.xlim(35,65);
    plt.ylim(0.5,2.5)
```

The halo bias $b_{h}$ satisfies $\delta_{h}=b_{h}\delta_{dm}$ and the galaxy bias $b_{g}$ satisfies $\delta_{g}=b_{g}\delta_{dm}$. This means that the correlation function of haloes $\xi_{h}=b_h^2\xi_{dm}$ and galaxies $\xi_{g}=b_g^2\xi_{dm}$. In the first two plots above, we are comparing the $\frac{\xi_{g}}{\xi_{h}}=\frac{b_g^2}{b_h^2}$ for different $\sigma_{HAM}$ in different distance. Unexpectingly, the ratios in small scales seem scale-free, but is noisy in large scales. The abnormal peak around 125 Mpc $h^{-1}$ is because their values are close to 0. Picking a relatively large scale 40-60 Mpc $h^{-1}$, we could see that the $\xi$ ratios are almost uniform there and larger $\sigma$ produce smaller signals. 

```python
fig,ax = plt.subplots()
x=np.array([0,0.1,0.2,0.31,0.4])
sel=(org['col1']>35)&(org['col1']>35)
y=np.array([1,np.mean((scat1['col2']/org['col2'])[sel]),np.mean((scat2['col2']/org['col2'])[sel]),np.mean((scat['col2']/org['col2'])[sel]),np.mean((scat3['col2']/org['col2'])[sel])])
ax.scatter(x,y,c='red')
plt.xlabel('$\sigma$')
plt.ylabel(' <$\\xi$ ratio>')
plt.title('2PCF monopole scatter/original');
plt.savefig('/global/cscratch1/sd/jiaxi/master/plots/sigma_diff.png',bbox_tight=True)
```

We plot the mean $\xi$ ratios for different $\sigma$ only in the 20Mpc $h^{-1}$ range and observe a descending trend as $\sigma$ becomes larger. This is because the larger the $\sigma$, the more likely a small haloes that are relatively far from other haloes will have an LRG inside. As a result, it will further weaken the positive correlation functions.
